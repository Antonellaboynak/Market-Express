```python
df_final2 = df_final2.fillna(value=fill_na_lan)
```

#### Clasificacíon de paises y sus tipos de centros

```python
cols_validar = ['Densidad_poblacional', 'LifeExpectancy', 'GNP', 'GNPOld', 'GNP_per_cápita']
df_final2[cols_validar] = df_final2[cols_validar].apply(pd.to_numeric, errors='coerce')
```

```python
## Condiciones para crear un centro único
condicion_centro_unico = (
    (df_final2['Densidad_poblacional'] > 409.86) &
    (df_final2['LifeExpectancy'] > 66) &
    (df_final2['IsOfficial'] == 'T') &
    (df_final2['Crecimiento_económico'] == 'Si') &
    (df_final2['GNP_per_cápita'] > 6089.69))

## Condición múltiples centros
condicion_multiples = (
    (df_final2['Densidad_poblacional'] > 25) &
    (df_final2['LifeExpectancy'] > 53) & 
    (df_final2['IsOfficial'] == 'T') &
    (df_final2['Crecimiento_económico'] == 'Si') &
    (df_final2['GNP_per_cápita'] > 500))
```

```python
condiciones = [condicion_centro_unico, condicion_multiples]
resultados = ['Centro Único', 'Múltiples Centros']
```

```python
df_final2['Recomendacion_Instalacion'] = np.select(
    condiciones, 
    resultados, 
    default='No se recomienda instalación')
```

```python
df_final2['Recomendacion_Instalacion'].value_counts()
```

```python
df_final2[(df_final2['Densidad_poblacional'] > 25) &
    (df_final2['LifeExpectancy'] > 53) &
    (df_final2['IsOfficial'] == 'T') &
    (df_final2['GNP'] > df_final2['GNPOld']) &
    (df_final2['GNP_per_cápita'] > 500)]
```

```python
df_final2['SurfaceArea'] = df_final2['SurfaceArea'].astype(int)
df_final2['IndepYear'] = pd.to_numeric(df_final2['IndepYear'], errors='coerce')
df_final2['IndepYear'] = df_final2['IndepYear'].astype('Int64')
df_final2['Population_country'] = df_final2['Population_country'].astype(int)
df_final2['ID'] = pd.to_numeric(df_final2['ID'], errors='coerce')
df_final2['ID'] = df_final2['ID'].astype('Int64')
df_final2['LifeExpectancy'] = df_final2['LifeExpectancy'].astype(float) 
df_final2['GNP'] = df_final2['GNP'].astype(float)  
df_final2['GNP_per_cápita'] = df_final2['GNP_per_cápita'].astype(float)
df_final2['Densidad_poblacional'] = df_final2['Densidad_poblacional'].astype(float) 
df_final2['Population_capital'] = pd.to_numeric(df_final2['Population_capital'], errors='coerce')
df_final2['Population_capital'] = df_final2['Population_capital'].astype('Int64')
```
